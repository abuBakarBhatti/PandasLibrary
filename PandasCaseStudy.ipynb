## Pandas Case Study
We are going to import libraries

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#importing Titanic dataset into our new variable name as kashti

kashti = sns.load_dataset("titanic")
kashti
```

### Printing first five rows of the dataset

```python
kashti.head(5)
```

### Using Describe Function

```python
kashti.describe()
```

### Conveting Data into CSV format

```python
kashti.to_csv("kashti.csv")
```

### Removing Data/columns from the data set

```python
new_Kashti = kashti.drop(["embark_town", "alone"], axis=1)
new_Kashti
```

### Mean of entire data frame

```python
kashti.mean()
```

### For Counting a particular column Values

```python
# using value_count function
kashti.value_counts(["survived"])
```

### To see the comparision we use GroupBy function

```python
kashti.groupby(["sex"]).mean()
```

#### We can also check the comparision between multiple columns

```python
kashti.groupby(["sex", "class", "who"]).mean()
```

### We can also use comparision operators to apply filters

```python
# We just want to see the chileren mean data in the sheet

kashti[kashti["age"] < 18].value_counts("age")

# above code shows filters only children and counts their ages 
```

### Using the above example and getting the mean 

```python
kashti[kashti["age"] < 18].mean()
```

```python
kashti[kashti["age"] < 18].groupby(["sex", "class"]).mean()
```
